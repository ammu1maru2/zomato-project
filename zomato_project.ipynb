ZOMATO PROJECT DATA **ANALYSIS


Step 1 *`*IMPORTING*`* THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2 Creating database

```python
dataframe = pd.read_csv("Zomato data .csv", encoding = 'latin-1')
print(dataframe)
```

CONVERT THE DATA TYPE OF COLUMN-RATE

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

SUMMARY OF DATAFRAME


```python
dataframe.info()
```

TYPES OF RESTURANT

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Types of restaurant")
```

CONCLUSION : THE MAJORITY OF THE RESTURANT FALLS INTO THE DINING CATERGORY
DINING RESTURANTS ARE PREFERRED BY LARGER NUMBER OF INDIVIDUALS


```python
dataframe.head()
```

THE MAJORITY OF RESTAURANT RECEIVED VOTES

```python
groupdata = dataframe.groupby(['listed_in(type)'])['votes'].sum()
result = pd.DataFrame({'votes':groupdata})
plt.plot(result, c="blue", marker="o")
plt.xlabel("types of restaurant", c="blue",size=20)
plt.ylabel("Votes", c="blue", size=20)
```

CONCLUSION - DINING RESTUARANTS RECEIVED MAXIMUM VOTES

THE MAJORITY OF RESTUARANTS RECEIVED RATINGS

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'])
plt.title("Rating Distribution")
plt.show()
```

CONCLUSION - THE MAJORITY RECEIVED RATINGS RANGING FROM 3.5 TO 4

WHICH MODES RECEIVES MAXIMUM RATING

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

CONCLUSION - OFFLINE RATINGS RECEIVED LOWER RATINGS IN COMPARISON TO ONLINE ORDERS, IN WHICH OBTAINED EXCELLENT RATING.

WHICH TYPE OF RESTUARANT RECEIVED MORE OFFLINE ORDER

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='count',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In(Type)")
plt.show()
```

CONCLUSION: DINING RESTUARANTS PRIMARILY ACCEPT OFFLINE ORDER, WHERAS CAFES PRIMARILY RECEIVE ONLINE ORDER.THE SUGGESTS THAT CLIENT PREFER TO PLACE ORDER IN PERSON AT RESTUARANTS ,BUT PERFER ONLINE ORDERING AT CAFES
